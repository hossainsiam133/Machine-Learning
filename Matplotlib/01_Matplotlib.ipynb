## Plotting

```python
# Plot With line and Without
import matplotlib.pyplot as plt
import numpy as np
xp = np.array([1,8])
yp = np.array([3,10])
plt.plot(xp,yp)
plt.show()

plt.plot(xp,yp,'o')
plt.show()
```

```python
# Multiple Points
xp = np.array([1,3,8,10])
yp = np.array([3,10,12,20])
plt.plot(xp,yp)
plt.show()
```

```python
# Default x axis: 0,1,2,3,...
yp = np.array([3,8,1,10,5,7])
plt.plot(yp)
plt.show()
```

## Markers

```python
yp = np.array([3,8,1,10,7])
plt.plot(yp,marker='x')
plt.show()
```

```python
yp = np.array([3,8,1,10,7])
# plt.plot(yp,'o:')
plt.plot(yp,marker='o',linestyle='-',color='b', ms=22, mfc='r',mec='y',linewidth=10)
plt.show()
```

## Line

```python
# Approach 1
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()

# Approach 2
x1 = np.array([0,1,2,3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0,1,2,3])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1,y1,x2,y2)
plt.show()
```

## Label

```python
xp = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
yp = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(xp,yp)

font1 = {'family':'serif','color':'red','size':20}
font2 = {'family':'serif','color':'blue','size':15}

plt.title("Sports Watch Data",loc='center',fontdict=font1)
plt.xlabel("Average Pulse",fontdict=font2)
plt.ylabel("Calorie Burnage",fontdict=font2)
plt.show()
```

##  Grid Lines

```python
xp = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
yp = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(xp,yp)
plt.grid()
# plt.grid(axis='x')
# plt.grid(axis='y')
plt.show()
```

## Multiple Plots

```python
import matplotlib.pyplot as plt
import numpy as np
xp = np.array([0,1,2,3])
yp = np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.title("Sales")
plt.plot(xp,yp)

xp = np.array([0,1,2,3])
yp = np.array([10, 20, 30, 40])
plt.subplot(1,2,2)
plt.title("Income")
plt.plot(xp,yp)

plt.suptitle("My Shop")
plt.show()
```

## Scatter

```python
xp = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
yp = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.subplot(1,2,1)
plt.title("Way 1")
plt.plot(xp,yp,'o')

plt.subplot(1,2,2)
plt.title("Way 2")
plt.scatter(xp,yp)
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = list(np.random.randint(150,size=13))
plt.scatter(x, y, c = 'red',s = sizes)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
sizes = list(np.random.randint(250,size=15))
colors = list(np.random.randint(100,size=15))
plt.scatter(x, y, c = colors, s = sizes)

plt.show()
```

## Bars

```python
import matplotlib.pyplot as plt
import numpy as np
name = np.array(["Siam","Talha","Asif","Mehedi","Alif"])
weight = np.array([65,80,55,70,90])
plt.bar(name,weight,width=0.5)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.barh(x,y,color = 'red',height=0.1)
plt.show()
```

## Histograms

```python
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

## Pie Charts


```python
portion = np.array([35,25,25,15])
plt.pie(portion)
plt.show()
# By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:
```

```python
portion = np.array([35,25,25,15])
myLabels = ["Apple","Mango","Guava","Pine"]

plt.pie(portion,labels=myLabels)
plt.show()
```

```python
portion = np.array([35,25,25,15])
myLabels = ["Apple","Mango","Guava","Pine"]

plt.pie(portion,labels=myLabels,startangle=90)
plt.show()
```

```python
portion = np.array([35,25,25,15])
myLabels = ["Apple","Mango","Guava","Pine"]
myExplode = np.array([0.1,0.1,0.1,0.1])

plt.pie(portion,labels=myLabels, explode=myExplode, shadow=True)
plt.legend(title='Fruits')
plt.show()
```
