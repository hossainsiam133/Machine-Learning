```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

slope,intercept,r,p,std_err = stats.linregress(x,y)

def func(x):
  return slope * x + intercept

plt.scatter(x,y)
model = list(map(func,x))
plt.plot(x,model,'r')
plt.scatter(10,slope*10+intercept)
plt.show()
print(func(10))
```

```python
import math

def fixed_point_iteration(g, x0, tol=1e-12, max_iter=10000):
    x = x0
    for i in range(max_iter):
        x_new = g(x)
        if abs(x_new - x) < tol:
            return x_new, i + 1
        x = x_new
    raise RuntimeError("Did not converge")

# Rearrangement: x = (1 - x^2)^(1/3)
g = lambda x: (1 - x**2) ** (1/3)

root, iters = fixed_point_iteration(g, x0=0.7)
print("Root ≈", root)
print("Iterations:", iters)
```
